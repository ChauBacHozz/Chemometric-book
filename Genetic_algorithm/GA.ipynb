```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

data = pd.read_excel("data.xlsx")
X = data.iloc[:, 1].values.reshape(-1, 1)
y = data.iloc[:, 2].values.reshape(-1, 1)

scaler1 = MinMaxScaler(feature_range=(-1,1))
X = scaler1.fit_transform(X)
scaler2 = MinMaxScaler(feature_range=(-1,1))
y = scaler2.fit_transform(y)
# scaler = 
# plt.plot(X, y, "ro")
# plt.show()
```

```python
def fitness(w):
    res = []
    for c in w:
        res.append(np.sum(np.square(y-(c * X)))/6)
    return np.array(res)
```

```python
populationSize = (100, 1)
population = np.random.uniform(low=-1, high=1,size=(populationSize[0], populationSize[1]))
```

```python
f1 = fitness(population)
```

```python
def selection(pop):
    f1 = fitness(population)
    pop = pop[np.argsort(f1)][:int(len(pop) * 0.2)]
    return pop
# selection(population)
```

```python
def crossover(p1, p2):
    return (p1 + p2) / 2
```

```python
def mutate(individual):
    new = individual + np.random.uniform(-0.1, 0.1)
    return new
```

```python
MUTATION_RATE = 0.5
EPOCHS = 10000
his = []
for i in range (EPOCHS):
    selected = selection(population)
    new_population = []
    while(len(new_population) < populationSize[0]):
        parent1 = selected[np.random.randint(0, len(selected))]
        parent2 = selected[np.random.randint(0, len(selected))]
        child = crossover(parent1, parent2)
        if np.random.random() < MUTATION_RATE:
            child = mutate(child)
        new_population.append(child)
    his.append(new_population[0])
```
