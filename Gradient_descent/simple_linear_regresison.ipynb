```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as MSE
import random

data = pd.read_excel("data.xlsx")
X = data.iloc[:, 1]
y = data.iloc[:, 2]

X = np.atleast_2d(X).reshape(-1, 1)
y = np.atleast_2d(y).reshape(-1, 1)
plt.plot(X, y, "ro")
plt.show()
```

```python
a = random.random()
b = random.random()
def d_weight(X, y, y_pred):
    return X.T @ (y_pred - y) / y_pred.shape[0]

def d_bias(y, y_pred):
    return np.mean(y_pred - y, axis=0)
```

```python
# Khai báo giá trị hệ số học máy
lr = 0.001
# Khai báo số bước lặp vừa đủ để thuật toán hội tụ
epochs = 10000
# Khai báo mảng loss để lưu trữ giá trị mất mát theo từng bước lặp
loss = []

for i in range (epochs):
    # Tính y dự đoán
    y_pred = a * X + b
    # Tính giá trị mất mát
    L = MSE(y, y_pred)
    loss.append(L)
    # Tính toán đạo hàm theo a và b
    da = d_weight(X, y, y_pred)
    db = d_bias(y, y_pred)
    # Cập nhật tham số a và b    
    a = a - da * lr
    b = b - db * lr
```

```python
plt.plot(loss)
plt.xlabel("Số bước lặp")
plt.ylabel("Giá trị mất mát")
plt.show()
```

```python
# Tạo điểm đầu cuối để vẽ đường hồi quy
X_line = np.atleast_1d((0, 50))
y_line = np.atleast_1d(a * X_line + b)[0]
# Vẽ đường hồi quy tuyến tính
plt.plot(X, y, "ro")
plt.plot(X_line, y_line, "b-")
plt.show()
```
