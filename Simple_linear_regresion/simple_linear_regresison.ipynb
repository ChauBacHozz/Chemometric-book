```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("data.xlsx")
X = data.iloc[:, 1]
y = data.iloc[:, 2]

plt.plot(X, y, "ro")
plt.show()
```

```python
# Tính giá trị trung bình của cột X và y 
X_mean = np.mean(X)
y_mean = np.mean(y)
# Tính độ lệch chuẩn của cột X và y
S_x = np.std(X)
S_y = np.std(y)
# Tính hệ số tương quan giữa x và y
r = np.corrcoef(X, y)[0, 1]**2
# In ra các giá trị vừa tính được
print(X_mean, y_mean, S_x, S_y, r)
```

```python
a = r * S_y / S_x
b = y_mean - a * X_mean
print(a, b)
```

```python
# Tạo điểm đầu cuối để vẽ đường hồi quy
X_line = np.atleast_1d((0, 50))
y_line = np.atleast_1d(a * X_line + b)
# Vẽ đường hồi quy tuyến tính
plt.plot(X, y, "ro")
plt.plot(X_line, y_line, "b-")
plt.show()
```

```python
def MSE(y, y_pred):
    n = y.shape[0]
    return np.sum(np.square(y - y_pred)) / n
```

```python
def MAE(y, y_pred):
    n = y.shape[0]
    return np.sum(np.abs(y - y_pred)) / n
```

```python
# Tính toán các giá trị diện tích dựa trên nồng độ
y_pred = a * X + b
print(MSE(y, y_pred))
```
