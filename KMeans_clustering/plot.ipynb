```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set a random seed for reproducibility
plt.figure(figsize=(6,6))
np.random.seed(42)
n_points = 100
# Class 1
class1_points = np.random.normal(loc=[4, 6], scale=[0.5, 0.5], size=(n_points, 2))
# Class 2
class2_points = np.random.normal(loc=[7, 6], scale=[0.5, 0.5], size=(n_points, 2))
# Class 3
class3_points = np.random.normal(loc=[6, 9], scale=[0.5, 0.5], size=(n_points, 2))
# Combine the points from both classes
all_points = np.concatenate([class1_points, class2_points, class3_points])
# Create labels for the classes (0 for class 1, 1 for class 2)
labels = np.array([0] * n_points + [1] * n_points + [2] * n_points)
# Scatter plot to visualize the dataset
plt.scatter(all_points[:, 0], all_points[:, 1], c=labels, cmap='viridis')
plt.xlabel('Đặc trưng 1')
plt.ylabel('Đặc trưng 2')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(all_points)
labels = kmeans.predict(all_points)
centroids = kmeans.cluster_centers_
centroids
```

```python
plt.figure(figsize=(6,6))
plt.scatter(all_points[:, 0], all_points[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='r', marker="^", s=200, edgecolors='k')
plt.xlabel('Đặc trưng 1')
plt.ylabel('Đặc trưng 2')
plt.show()
```
